```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Cargar el conjunto de datos
ruta_archivo = "C:/Users/Usuario/Documents/Henry/PI/Dataset/movies_dataset_modificado.csv"
dataset = pd.read_csv(ruta_archivo)

# Vista general del conjunto de datos
print(dataset.head())  # Muestra las primeras filas del conjunto de datos
print(dataset.shape)  # Muestra la cantidad de filas y columnas del conjunto de datos
print(dataset.describe())  # Estad√≠sticas descriptivas de las columnas num√©ricas

# An√°lisis de variables individuales
# Histograma de la variable 'vote_average'
sns.histplot(data=dataset, x='vote_average', bins=20)
plt.title('Distribuci√≥n de las puntuaciones promedio de las pel√≠culas')
plt.xlabel('Puntuaci√≥n promedio')
plt.ylabel('Frecuencia')
plt.show()

# Gr√°fico de barras de los g√©neros m√°s frecuentes
genres_count = dataset['genres'].value_counts().head(10)
plt.figure(figsize=(10, 6))
sns.barplot(x=genres_count.values, y=genres_count.index)
plt.title('G√©neros m√°s frecuentes')
plt.xlabel('Cantidad')
plt.ylabel('G√©nero')
plt.show()

# Matriz de correlaci√≥n entre variables num√©ricas
numeric_columns = ['budget', 'revenue', 'runtime', 'vote_average', 'vote_count']
correlation_matrix = dataset[numeric_columns].corr()
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Matriz de correlaci√≥n')
plt.show()

# An√°lisis de valores faltantes
missing_values = dataset.isnull().sum()
plt.figure(figsize=(10, 6))
sns.barplot(x=missing_values.values, y=missing_values.index)
plt.title('Valores faltantes en el conjunto de datos')
plt.xlabel('Cantidad de valores faltantes')
plt.ylabel('Variable')
plt.show()

```
